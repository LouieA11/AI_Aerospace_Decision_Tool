# AI and Aerospace Explorer

This notebook is a stronger research project with:
- a table of aerospace AI applications
- scoring and ranking
- a filterable interactive view
- a chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML, Markdown
import ipywidgets as widgets
plt.rcParams["figure.figsize"] = (7,4)
```

```python
apps = pd.DataFrame({
    "Application":["Satellite image analysis","Aircraft fault detection","Weather forecasting","Autonomous navigation","Space mission planning"],
    "Impact":[9,8,9,8,7],
    "Risk":[4,5,4,7,6],
    "Complexity":[7,8,8,9,8]
})
apps["Usefulness"] = apps["Impact"]*2 - apps["Risk"] + apps["Complexity"]*0.5
apps
```

```python
app_choice = widgets.Dropdown(options=list(apps["Application"]), description="App")
button = widgets.Button(description="Show App")
out = widgets.Output()

def show_app(_):
    with out:
        out.clear_output()
        row = apps[apps["Application"] == app_choice.value].iloc[0]
        print("Application:", row["Application"])
        print("Impact:", row["Impact"])
        print("Risk:", row["Risk"])
        print("Complexity:", row["Complexity"])
        print("Usefulness score:", row["Usefulness"])
        fig, ax = plt.subplots()
        ax.bar(["Impact","Risk","Complexity"], [row["Impact"], row["Risk"], row["Complexity"]])
        ax.set_title("Aerospace AI Snapshot")
        plt.show()

button.on_click(show_app)
display(app_choice, button, out)
```
