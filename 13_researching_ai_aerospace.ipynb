# AI and Aerospace Research Starter

This project helps students organize and analyze a few aerospace uses of AI.

Possible examples:
- navigation
- weather prediction
- fault detection
- mission planning
- satellite image analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Step 1: Create an aerospace AI table

```python
apps = pd.DataFrame({
    "Application": [
        "Satellite image analysis",
        "Aircraft fault detection",
        "Weather forecasting",
        "Autonomous navigation",
        "Space mission planning"
    ],
    "ImpactScore": [9, 8, 9, 8, 7],
    "RiskScore": [4, 5, 4, 7, 6],
    "ComplexityScore": [7, 8, 8, 9, 8]
})
apps
```

## Step 2: Compute a simple usefulness score

```python
apps["UsefulnessScore"] = apps["ImpactScore"] * 2 - apps["RiskScore"] + apps["ComplexityScore"] * 0.5
apps
```

## Step 3: Rank the applications

```python
apps.sort_values("UsefulnessScore", ascending=False)
```

## Step 4: Visualize the applications

```python
plt.figure(figsize=(7,4))
plt.bar(apps["Application"], apps["UsefulnessScore"])
plt.xticks(rotation=30, ha="right")
plt.title("Estimated Aerospace AI Usefulness")
plt.ylabel("Score")
plt.tight_layout()
plt.show()
```
